```python
x = np.array([1, 2])
y = np.array([-2, 1])

## 2*2 Matrix
```

```python
## Dot Product
a = np.dot(x, y)
print(a)
print(x.T @ y)
```

```python
## Normalise the vector(Find the unit vector: nor(x(a,b)) = sq(a^2 + b^2)) or Finding the determinant
b = np.linalg.norm(x)
## Using the standard functions
c = np.sqrt(x[0]**2 + x[1]**2)
print(b, c)
```

```python
## Find the angle of the vector
theta = np.arccos(np.dot(x,y) / (np.linalg.norm(x) * np.linalg.norm(y))) ## Correction in the lab1 Formula
print(theta*180/np.pi)
```

```python
# Uniform Distribution
x = np.random.rand(1000,1)
y = np.random.rand(1000, 1)
z = np.random.rand(100000000,1)
fig, ax = plt.subplots(nrows=1, ncols=3,figsize=(14,4))
n1bins, n2bins, n3bins = 4, 40, 100 ## Equal width bin in a range

ax[0].hist(x, bins=n1bins)
ax[0].set_ylim(0,300)
ax[0].set_xlabel("Bins", fontsize=12)
ax[0].set_ylabel("Count", fontsize=12)
ax[0].tick_params(axis='both', which='major', labelsize=14)
ax[0].set_title("(a) Random No=%4d"%(x.size) + " Bins=%4d" %(n1bins), fontsize=12)

ax[1].hist(y, bins=n2bins)
ax[1].set_ylim(0,50)
ax[1].set_xlabel("Bins", fontsize=12)
ax[1].set_ylabel("Count", fontsize=12)
ax[1].tick_params(axis='both', which='major', labelsize=12)
ax[1].set_title("(b) Random No=%4d"%(y.size) + " Bins=%4d" %(n2bins), fontsize=12)

ax[2].hist(z, bins=n3bins)
ax[2].set_ylim(0,1100000)
ax[2].set_xlabel("Bins", fontsize=12)
ax[2].set_ylabel("Count", fontsize=12)
ax[2].tick_params(axis='both', which='major', labelsize=12)
ax[2].set_title("(c) Random No=%4d"%(z.size) + " Bins=%4d" %(n3bins), fontsize=9)

plt.savefig("histograms_uniform.png", bbox_inches="tight")
plt.tight_layout(pad = 5.0)
```

```python
#Normal Distribution
N = 1000
M = 100000
x1 = np.zeros(N)
x2= np.zeros(M) 
for n in range(N):
    x1[n] = np.sum(np.random.rand(20,1)) - np.sum(np.random.rand(10,1))

for n in range(M):
    x2[n] = np.sum(np.random.rand(10000,1)) - np.sum(np.random.rand(1000,1))
    
fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(10,6))
ax[0].set_xlabel("Bins", fontsize=12)
ax[0].set_ylabel("Count", fontsize=12)
ax[0].set_title("(a) Random No=%4d"%(N) +" x=20 y=10" + " Bins=20", fontsize=9)
ax[0].hist(x1, 20)

ax[1].set_xlabel("Bins", fontsize=12)
ax[1].set_ylabel("Count", fontsize=12)
ax[1].set_title("(b) Random No=%4d"%(M) +" x=10000 y=10"+" Bins=1000", fontsize=9)
ax[1].hist(x2, 20)

plt.savefig("normal_distribution.png", bbox_inches="tight")
plt.tight_layout(pad = 5.0)
```

```python
MaxTrial = 2000
sampleSizeRange = np.linspace(100, 1500, 1000)
plotVar = np.zeros(len(sampleSizeRange))
for sSize in range(len(sampleSizeRange)):
    numSamples = int(sampleSizeRange[sSize])
    vStrial = np.zeros(MaxTrial)
    for trial in range(MaxTrial):
        xx = np.random.randn(numSamples,1)
        vStrial[trial] = np.var(xx)
    plotVar[sSize] = np.var(vStrial)
fig, ax = plt.subplots(figsize=(4,4))
ax.set_xlabel("Sample Size", fontsize=12)
ax.set_ylabel("Variance", fontsize=12)
ax.set_title("Sample Range: 100 to 1500 Sample Size: 1000", fontsize=9)
ax.plot((plotVar))
plt.savefig("Uncertainity_Estimation2.png", bbox_inches="tight")
```

```python
C = np.array([[2.0,1.0], [1.0,2.0]])
A = np.linalg.cholesky(C)
print(A)
print(A @ A.T)
```

```python
X = np.random.randn(2000,2)
print(X)
Y = X @ A
print(X.shape)
print(Y.shape)
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(Y[:,0], Y[:,1], c="r", s=4)
ax.scatter(X[:,0], X[:,1], c="y", s=4)
ax.set_xlim(-6, 6)
ax.set_ylim(-6, 6)
ax.set_title("Scatter of Isotropic and Correlated Gaussian Densities", fontsize=12)

plt.savefig("SamplingMultiVariant.png", bbox_inches="tight")
```

```python
nPoints = 50
pVars = np.zeros(nPoints)
thRange = np.linspace(0, 2*np.pi, nPoints)
for n in range(nPoints):
    theta = thRange[n]
    u = [np.sin(theta), np.cos(theta)]
    pVars[n] = np.var(Y @ u)
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(pVars)
ax.set_title("Distribution of Projections", fontsize=12)
ax.set_xlabel("Theta", fontsize=12)
ax.set_ylabel("Variance", fontsize=12)

plt.savefig("distributionOfProjections.png", bbox_inches="tight")
```
